```python
# Dependencies and Setup
import pandas as pd
import gmaps
import requests


# Import API key
from api_keys import gkey
```

```python
gmaps.configure(api_key=gkey)
```

```python
jhcounty_cases = pd.read_csv('Cases by day by county.csv')
jhcounty_cases
```

```python
locations = jhcounty_cases[["Lat","Long_"]].astype(float)

cases = jhcounty_cases["6/17/2020"].astype(float)

fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights = cases,
                                 dissipating=False, max_intensity=100,
                                 point_radius = .2)

fig.add_layer(heat_layer)

fig
```

```python
low_cases_tx = jhcounty_cases.loc[(jhcounty_cases['Province_State']=='Texas') & (jhcounty_cases['6/17/2020']<10)]
low_cases_tx
```

```python
hotel_df = pd.DataFrame(low_cases_tx)
hotel_df.reset_index(drop=True, inplace = True)
hotel_df = hotel_df.drop([83,84])
hotel_df = hotel_df[['Admin2','Province_State','Lat','Long_','6/17/2020']]
hotel_df['Hotel Name'] = ''
```

```python
params = {
    "radius": 2000,
    "types": "lodging",
    "keyword": "hotel",
    "key" : gkey}

#query for lodging/hotels within 2000 meter radius, if no hotel found within these parameters, populating DF with "No Hotel"
for index, row in hotel_df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    lat=row['Lat']
    lng=row['Long_']
    
    params["location"] = f"{lat},{lng}"
    
    hotel_query = requests.get(base_url, params=params).json()
    
    try:
        hotel_df.loc[index, "Hotel Name"] = hotel_query["results"][0]["name"]
    except (KeyError,IndexError):
        print('Missing field/result...populating "No Hotel".')
        hotel_df.loc[index, "Hotel Name"] = "No Hotel"
hotel_df
```

```python
hotel_df.to_csv('covid hotels.csv')
```

```python
hotel_df = pd.read_csv('covid hotels.csv')
hotels_df = hotel_df.loc[hotel_df['Hotel Name'] != 'No Hotel']
hotels_df
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content = hotels_info)

# Display figure
fig.add_layer(markers)
fig
```
